```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel (r'result.xlsx') #for an earlier version of Excel, you may need to use the file extension of 'xls'
df.set_index('times', drop=True, append=False, inplace=True, verify_integrity=False)
print(df)
```

```python
plt.figure(figsize=(9,6))
# df.plot(kind="line")
plt.plot(df["Belady"], color='g')
plt.plot(df["LFU"], color='r')
plt.plot(df["LRU"], color='#FFCD63')
plt.plot(df["Our policy"], color = '#A91188')
plt.xlabel("Requst times",fontsize = 14)
plt.ylabel("Hit rate",fontsize = 14)
plt.title("Experiment: train_casa09_predict_casa12_C100",fontsize = 14)
plt.legend()
plt.savefig('ML model result for storage system.png')
```

```python
hit_rate = [0.152458, 0.128627, 0.068589, 0.03606070741700811]
methods = ['Belady','LFU','LRU','Our policy']

plt.figure(figsize=(9,6))

plt.bar(methods,hit_rate,color='b',label='Average')


plt.xlabel("Cache Replacement Methods",fontsize = 14)
plt.ylabel("Final Hit rate",fontsize = 14)
plt.title("Experiment: train_cheetah12_predict_casa09_C100",fontsize = 14)
plt.legend()
# plt.savefig('ML model result for storage system.png')
```
